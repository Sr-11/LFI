# Kernel
modify n_train and path_LFI to change the checkpoint.

```python
n_train = 1300101
path_LFI = './checkpoint1300000/30/'
model = DN(300, 100).to(device)
another_model = another_DN().to(device)
model,another_model,epsilonOPT,sigmaOPT,sigma0OPT,cst = load_model(model, another_model, path_LFI)
```

```python
n_eval = 10000
X_eval = dataset_P[np.random.choice(n_train, n_eval, replace=False)]
Y_eval = dataset_Q[np.random.choice(n_train, n_eval, replace=False)]
thres, sig_to_sig, back_to_back = get_thres_at_once(X_eval, Y_eval, X_eval, Y_eval,
                    model, another_model, epsilonOPT, sigmaOPT, sigma0OPT, cst, 
                    batch_size = 10000)
thres, sig_to_sig, back_to_back
```

```python
n_test = 1000000
X_test = dataset_P[n_train+n_eval : n_train+n_eval+n_test]
Y_test = dataset_Q[n_train+n_eval : n_train+n_eval+n_test]
XY_test = torch.concatenate((X_test, Y_test), axis=0)
phi_Z = torch.zeros(2*n_test, dtype=dtype, device=device)
batch_size = 10000
for i in range(n_test//batch_size):
    phi_Z[i*batch_size:(i+1)*batch_size] = compute_score_func(XY_test[i*batch_size:(i+1)*batch_size], X_eval, Y_eval, 
                                                            model, another_model, epsilonOPT, sigmaOPT, sigma0OPT, cst)
idx = phi_Z > thres
```

```python

feature = 26
Background = X_test[:, feature].cpu().numpy()
Signal = Y_test[:, feature].cpu().numpy()
Conditioned_to_signal = XY_test[idx, feature].cpu().numpy()
Conditioned_to_background = XY_test[~idx, feature].cpu().numpy()
bins = 500
alpha = 1
plt.figure(figsize=(5, 4), dpi=100)
plt.hist(Background, bins=bins, alpha=alpha, label='Background', density=True, color='black', histtype='step')
plt.hist(Signal, bins=bins, alpha=alpha, label='Signal', density=True, color='red', histtype='step')
plt.hist(Conditioned_to_signal, bins=bins, alpha=alpha, label='Above threshold', density=True, color='blue', histtype='step')
plt.hist(Conditioned_to_background, bins=bins, alpha=alpha, label='Below threshold', density=True, color='green', histtype='step')
plt.xlabel('$m_{Wbb}$(normalized)')
plt.ylabel('Frequenc of events')
plt.xlim(0.4, 1.6)
plt.legend(loc='upper right', fontsize=10)
plt.tight_layout()
plt.savefig('./paper/kernel m_wbb.pdf', dpi=300)
```
