```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Default Dataset.csv', header=None)
```

```python
plt.plot(df[0], df[1], 'o')
```

```python
p_list = np.zeros(df[0].shape[0])
for i, (a,b) in enumerate(zip(df[0], df[1])): 
    if a>1 or b>1:
        continue
    p_list[i] = 1/11 * np.sqrt(1100) * (a+b-1)/np.sqrt(b*(1-b))
```

```python
plt.plot(p_list)
```
