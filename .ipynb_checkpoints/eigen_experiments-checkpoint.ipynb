```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def K(x,y):
    return 1/np.sqrt(2*np.pi) * np.exp(-(x-y)**2)
```

```python
n = 100;
Kxy = np.zeros((n,n))
for i in range(n):
    for j in range(n):
        Kxy[i,j] = K(i/(n-1),j/(n-1))
```

```python
w,v = np.linalg.eigh(Kxy)
idx = np.argsort(-w)
w = w[idx]/n
v = v[:,idx]*np.sqrt(n)
```

```python
plt.plot(np.arange(n)/n, v[:,0:3])
```

```python
L1 = np.sum(np.abs(v),axis=0)/n
L2 = np.sum(v*v,axis=0)/n
```

```python
plt.scatter(range(n),w,label='lambda')
plt.scatter(range(n),L1,label='L1')
plt.scatter(range(n),L2,label='L2')
plt.legend()
```

```python
import numpy as np
np.all(np.linalg.eigvals(Kxy) > 0)
```

```python
plt.matshow(np.real(v@v.T)/n)
plt.colorbar()
```
